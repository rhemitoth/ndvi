```python
def open_clean_band(band_path, crop_layer=None):
    """A function that opens a Landsat band as an (rio)xarray object

    Parameters
    ----------
    band_path : list
        A list of paths to the tif files that you wish to combine.

    crop_layer : geopandas geodataframe
        A geodataframe containing the clip extent of interest. NOTE: this will 
        fail if the clip extent is in a different CRS than the raster data.

    Returns
    -------
    An single xarray object with the Landsat band data.

    """

    if crop_layer is not None:
        try:
            clip_bound = crop_layer.geometry
            cleaned_band = rxr.open_rasterio(band_path,
                                             masked=True).rio.clip(clip_bound,
                                                                   from_disk=True).squeeze()
        except Exception as err:
            print("Oops, I need a geodataframe object for this to work.")
            print(err)
    else:
        cleaned_band = rxr.open_rasterio(band_path,
                                         masked=True).squeeze()

    return cleaned_band
```

```python
#Grouping bands

b4 = [b4_20140722,
     b4_20140807,
     b4_20140823,
     b4_20140908,
     b4_20140924]

b5 = [b5_20140722,
     b5_20140807,
     b5_20140823,
     b5_20140908,
     b5_20140924]

qa = [qa_20140722,
     qa_20140807,
     qa_20140823,
     qa_20140908,
     qa_20140924]

dates = ['20140722',
     '20140807',
     '20140823',
     '20140908',
        '20140924']

files = [files_20140722,
        files_20140807,
        files_20140823,
        files_20140908,
        files_20140924]
```

```python
#Calculating NDVI

num_images = len(b4)
ndvi_files = []

for i in range(num_images):
    
    red = b4[i]
    nir = b5[i]
    
    
    #red = (red * 0.0000275) + -0.2
    #nir = (nir * 0.0000275) + -0.2
    ndvi_calc = np.where((nir+red)==0., 0, (nir-red)/(nir+red))
    
    
    filename = '/Users/rhemitoth/Documents/Lion_Movement/R/hip_lion_issa/data_processed/NDVI/from_python/' + 'ndvi' + '%s.tiff' % dates[i]
    
    im = rasterio.open(files[i][1])
    
    ndviImage = rasterio.open(filename,'w',driver='Gtiff',
                          width= im.width, 
                          height = im.height, 
                          count=1, crs= im.crs, 
                          transform= im.transform, 
                          dtype='float64')
    ndviImage.write(ndvi_calc,1)
    ndviImage.close()
    

    ndvi_files.append(filename)
        
```

```python
#Reading in NDVI images

ndvi = []

for i in range(len(ndvi_files)):
    image = rxr.open_rasterio(ndvi_files[i])
    ndvi.append(image)
```

```python
# Masking Rasters

for i in range(len(ndvi_files)):
    qaBand = qa[i]
    file = rasterio.open(ndvi_files[i])
    open_file = file.read(1).astype('float64')
    cloud_mask = unpackqa.unpack_to_array(qaBand, 
                                        product='LANDSAT_8_C2_L2_QAPixel', 
                                        flags=['Cloud'])
    cloud_shadow_mask = unpackqa.unpack_to_array(qaBand, 
                                        product='LANDSAT_8_C2_L2_QAPixel', 
                                        flags=['Cloud_Shadow'])
    mask = np.ma.mask_or(cloud_mask, cloud_shadow_mask)
    ndvi_masked = np.where(mask == 1, np.nan, open_file)
    
    quick_raster_plot(ndvi_masked)
    filename = '/Users/rhemitoth/Documents/Lion_Movement/R/hip_lion_issa/data_processed/NDVI/from_python/' + 'ndvi' + '%s.tiff' % dates[i]
    
    im = rasterio.open(files[i][1])
    
    ndviImage = rasterio.open(filename,'w',driver='Gtiff',
                              width= im.width, 
                              height = im.height, 
                              count=1, crs= im.crs,
                              transform= im.transform, 
                              dtype='float64',
                             nodata = -99)
    ndviImage.write(ndvi_masked,1)
    ndviImage.close()
```

```python
#Reading in NDVI images

ndvi = []

for i in range(len(ndvi_files)):
    image = rxr.open_rasterio(ndvi_files[i])
    ndvi.append(image)

#Updating CRS
for i in range(len(ndvi)):
    crs_36S = CRS.from_string('EPSG:32736')
    ndvi[i] = ndvi[i].rio.reproject(crs_36S)
    ndvi[i].rio.to_raster("%s" %ndvi_files[i])
```
